```python
import pandas as pd 
```

```python
songs_normalize = pd.read_csv('songs_normalize.csv')
```

```python
print(songs_normalize.isna())
```

```python
songs_normalize = songs_normalize.drop(columns=['speechiness','acousticness','instrumentalness','liveness','valence','mode','explicit','duration_ms','genre'])
display(songs_normalize.head(50))
```

```python
songs_normalize=songs_normalize[songs_normalize['year']>=2000]
```

```python
songs_normalize=songs_normalize[['song','artist','popularity','year','danceability','energy','loudness','tempo','key']]
```

```python
songs_normalize=songs_normalize.rename(columns={'year':'release_year'})
```

```python
songs_normalize = songs_normalize.reset_index(drop=True)
```

```python
duplicate_records = songs_normalize[songs_normalize.duplicated(keep=False)]

# keep=False will identify and display every row that has duplicates, ensuring that all instances of the duplicated data are shown
print(f"total records: {len(songs_normalize)} \n")
print(f"duplicate records: {len(songs_normalize[songs_normalize.duplicated(keep='first')])} \n")
print("All instances of the duplicated data","\n", "-"*50)
display(duplicate_records)
```

```python
# remove duplicates
songs_normalize = songs_normalize.drop_duplicates()

# recheck for duplicates after dropping
duplicate_records = songs_normalize[songs_normalize.duplicated(keep=False)]

print(f"total records: {len(songs_normalize)} \n")
print(f"duplicate records: {len(songs_normalize[songs_normalize.duplicated(keep='first')])} \n")
print("All instances of the duplicated data","\n", "-"*50)
display(duplicate_records)
```

```python
duplicates = songs_normalize[songs_normalize.duplicated(subset=['song', 'artist','release_year'], keep=False)]
display(duplicates)
display(f"amount of duplicates:{len(duplicates)}")
```

```python
songs_normalize = songs_normalize.sort_values('popularity', ascending=False).drop_duplicates(subset=['song', 'artist', 'release_year'], keep='first')
```
