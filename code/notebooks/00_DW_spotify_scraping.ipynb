```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

# Replace with your Spotify Client ID and Secret
client_id = 'de3a68a4d1f04d98b87938df85167b4f'
client_secret = '1dbf52d4cf5c4776b7e36b399906507e'

# Set up the client credentials
auth_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
sp = spotipy.Spotify(auth_manager=auth_manager)
```

```python
# Define the playlist ID
playlist_id = '3pKNWa2JWhnNv05Gu5c9p7'

track_data = []

# Initialize the offset to fetch tracks in batches
offset = 0
limit = 100  

while True:
    results = sp.playlist_tracks(playlist_id, offset=offset, limit=limit)

    # Loop through the tracks to extract relevant data
    for item in results['items']:
        track = item['track']
        track_info = {
            'name': track['name'],
            'artist': track['artists'][0]['name'],
            'popularity': track['popularity'],
            'release_year': track['album']['release_date'][:4],
            'track_id': track['id']
        }
        track_data.append(track_info)

    if len(results['items']) < limit:
        break  

    # Increment the offset to fetch the next batch of tracks
    offset += limit

# DF
import pandas as pd
track_df = pd.DataFrame(track_data)

# Display the DataFrame
display(track_df)
```

```python
# Get audio features for each track
track_features_data = []

for track in track_data:
    track_id = track['track_id']
    audio_features = sp.audio_features(track_id)  # Fetch audio features for each track

    if audio_features[0]:  # Check if audio features are available
        features = audio_features[0]
        track_features_data.append({
            'name': track['name'],
            'artist': track['artist'],
            'popularity': track['popularity'],
            'release_year': track['release_year'],
            'danceability': features['danceability'],
            'energy': features['energy'],
            'loudness': features['loudness'],
            'tempo': features['tempo'],
            'key': features['key']
        })

# Convert to DataFrame for easier manipulation
track_features_df = pd.DataFrame(track_features_data)

# Display the DataFrame with audio features
display(track_features_df)
```

```python
#Save as a csv
track_features_df.to_csv('raw_spotify.csv', index=False)
```
